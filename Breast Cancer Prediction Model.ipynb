Handling Null Values

```python
dataset.iloc[:,18]=dataset['radius_se'].fillna(value=dataset['radius_se'].mean())
dataset.iloc[:,12]=dataset['radius_se'].fillna(value=dataset['radius_se'].mean())
dataset.iloc[:,11]=dataset['fractal_dimension_mean'].fillna(value=dataset['fractal_dimension_mean'].mean())
```

```python
dataset=dataset.drop(['id'], axis=1)
```

```python
import seaborn as sns
plt.figure(figsize=(20,10))
cor = dataset.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr = set()  
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]  
                col_corr.add(colname)
    return col_corr
```

```python
dataset.head(5)
```

```python
x=dataset.iloc[:, 1:].values
y=dataset.iloc[:, 0].values

```

```python
corr_features = correlation(dataset, 0.7)
print(corr_features)
```

```python
dataset.drop(corr_features,axis=1,inplace=True)
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train[:,:]=sc.fit_transform(x_train[:,:])
x_test[:,:]=sc.transform(x_test[:,:])
print(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
```

Using DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
log=DecisionTreeClassifier()
log.fit(x_train,y_train)
y_test_pred=log.predict(x_test)
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
print('Accuracy :',accuracy_score(y_test,y_test_pred))
print('Precision :',recall_score(y_test,y_test_pred))
print('Recall :',recall_score(y_test,y_test_pred))
print('F1 score:',f1_score(y_test,y_test_pred))
```

Using Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
log=GaussianNB()
log.fit(x_train,y_train)
y_test_pred=log.predict(x_test)
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
print('Accuracy :',accuracy_score(y_test,y_test_pred))
print('Precision :',recall_score(y_test,y_test_pred))
print('Recall :',recall_score(y_test,y_test_pred))
print('F1 score:',f1_score(y_test,y_test_pred))
```

```python
import pandas as pd
dataset = pd.read_csv('driver_profile_data.csv')
dataset.drop(['PathOrder'],axis=1,inplace=True)
```

```python
#the 'Class' was categorical value, so we are encoding the fearture.
k=[i for i in range(1,11)]
dataset['Class'].replace(['A','B','C','D','E','F','G','H','I','J'],k,inplace=True)
```

```python
y=dataset.iloc[:,-1].values
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataset.drop(['Class'], axis=1),y, test_size = 0.2,random_state=0)
```

```python
from sklearn.feature_selection import mutual_info_classif
mutual_info = mutual_info_classif(x_train, y_train)
mutual_info = pd.Series(mutual_info)
mutual_info.index = x_train.columns
mutual_info.sort_values(ascending=False)
```

```python
from sklearn.feature_selection import SelectKBest
sel_five_cols = SelectKBest(mutual_info_classif, k=25)
sel_five_cols.fit(x_train, y_train)
x_train.columns[sel_five_cols.get_support()]
```

```python
remo=['Fuel_consumption', 'Accelerator_Pedal_value',
       'Throttle_position_signal', 'Intake_air_pressure',
       'Absolute_throttle_position', 'Engine_soacking_time',
       'Long_Term_Fuel_Trim_Bank1', 'Engine_torque_after_correction',
       'Torque_of_friction', 'Flywheel_torque_(after_torque_interventions)',
       'Engine_coolant_temperature', 'Engine_Idel_Target_Speed',
       'Engine_torque', 'Calculated_LOAD_value',
       'Maximum_indicated_engine_torque', 'Flywheel_torque',
       'Activation_of_Air_compressor', 'Torque_converter_speed',
       'Engine_coolant_temperature.1', 'Wheel_velocity_rear_right-hand',
       'Wheel_velocity_front_right-hand', 'Wheel_velocity_rear_left-hand',
       'Torque_converter_turbine_speed_-_Unfiltered',
       'Master_cylinder_pressure', 'Steering_wheel_angle']
```

```python
#Dimensionality Reduction
col=list(dataset.columns)
for i in remo:
    col.remove(i)
col.remove('Class')
```

```python
dataset.drop(col,axis=1,inplace=True)
dataset.head()
```

```python
target=pd.get_dummies(dataset['Class'])
target
```

```python
dataset.drop(['Class'],axis=1,inplace=True)
dataset
```

```python
dataset=dataset.join(target,how='right')
```

```python
dataset
```

```python
x = dataset.iloc[:,0:-10].values
y=dataset.iloc[:,25:35].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
classifier.add(Dense(15, activation = 'sigmoid', input_dim = 25))
classifier.add(Dense(15, activation = 'sigmoid'))
classifier.add(Dense(12, activation = 'sigmoid'))
classifier.add(Dense(10, activation = 'softmax'))
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
classifier.fit(x_train, y_train, batch_size = 10,epochs= 60)
```

```python
score=classifier.evaluate(x_test,y_test,verbose=0)
print("Loss: ",score[0])
print('Accuracy: ',score[1])
```
